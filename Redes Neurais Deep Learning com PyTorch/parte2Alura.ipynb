**CAMADA LINEAR**


```python
import torch
from torch import nn

torch.manual_seed(42)

perceptron = nn.Linear(3, 1)

for nome, tensor in perceptron.named_parameters():
  print(nome, tensor.data)

w1, w2, w3 = perceptron.weight.data.numpy()[0]
b = perceptron.bias.data.numpy()

#w1 * x1 + w2 * x2 + w3 * x3 + b = 0
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot3d(perceptron):

  w1, w2, w3 = perceptron.weight.data.numpy()[0]
  b = perceptron.bias.data.numpy()

  X1 = np.linspace(-1,1,10)
  X2 = np.linspace(-1,1,10)

  X1,X2 = np.meshgrid(X1,X2)

  X3 = (b - w1 * X1 - w2 * X2) / w3

  fig = plt.figure()
  ax = fig.add_subplot(projection = '3d')
  ax.view_init(azim=90)

  surf = ax.plot_surface(X1,X2,X3 , cmap = 'plasma')

#a,b,c,d = 1,2,3,4
```

```python
plot3d(perceptron)
```

```python
X = torch.Tensor([0,1,2])
y = perceptron(X)

plot3d(perceptron)
plt.plot([X[0]], [X[1]], [X[2]], marker = '^', markersize = 15)
```
